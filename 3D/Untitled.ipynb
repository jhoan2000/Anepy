```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import numpy as np
ax = plt.figure(figsize=(8,6)).add_subplot(111, projection='3d')
x = np.array([1.5,5])
y = np.array([[10.5,6]])
z = np.array([[5,7]])

ax.scatter3D(x, y, z)

ax.set_title("3d axes in Matplotlib",fontsize=14,fontweight="bold")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")


plt.show()
```
